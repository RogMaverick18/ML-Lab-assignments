```python
# IDCIA Complete: Segmentation (Enhanced U-Net++) + Classification on Segmented Cells

# ✅ Setup & Imports
!pip install medpy
!pip install seg-metrics
import os
import numpy as np
import pandas as pd
import cv2
from glob import glob
import matplotlib.pyplot as plt
from PIL import Image
from skimage.draw import disk
from scipy import ndimage

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader, random_split
import torchvision.transforms as transforms
import torchvision.models as models
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
from scipy.spatial.distance import directed_hausdorff
import torch.nn.functional as F
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# Mixed precision imports
from torch.cuda.amp import GradScaler, autocast

os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# Paths
IDCIA_PATH = "/kaggle/input/idcia-dataset"
OUTPUT_PATH = "/kaggle/working"
MODEL_PATH = "/kaggle/input/idcia-dataset"

# ✅ Memory-optimized constants
SEED = 42
BATCH_SIZE = 4
BASE_CHANNELS = 48
NUM_EPOCHS = 20
SEG_LEARNING_RATE = 1e-5  # Reduced learning rate for fine-tuning
CLS_LEARNING_RATE = 0.001  # Learning rate for classification
USE_MIXED_PRECISION = True
IMG_SIZE = 512
DISK_RADIUS = 5
USE_TTA = True  # Use test-time augmentation
USE_ATTENTION = True  # Use attention gates
LOAD_PRETRAINED = False  # Set to False to train from scratch

# Set random seed for reproducibility
def seed_everything(seed):
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True

seed_everything(SEED)

# ✅ Visualize Sample Image with Dots
def plot_sample_with_dots(image_path, dots_csv_path):
    img = Image.open(image_path)
    img_np = np.array(img)
    dots = pd.read_csv(dots_csv_path)
    plt.figure(figsize=(8,6))
    plt.imshow(img_np, cmap='gray')
    plt.scatter(dots['X'], dots['Y'], color='red', s=10)
    plt.title(os.path.basename(image_path))
    plt.axis('off')
    plt.show()

# Example usage:
nestin_img_dir = os.path.join(IDCIA_PATH, 'IDCIA/images/Nestin')
nestin_csv_dir = os.path.join(IDCIA_PATH, 'IDCIA/ground_truth/Nestin')
nestin_imgs = sorted(glob(os.path.join(nestin_img_dir, '*.tiff')))
nestin_csvs = sorted(glob(os.path.join(nestin_csv_dir, '*.csv')))

if nestin_imgs and nestin_csvs:
    first_image_name = os.path.splitext(os.path.basename(nestin_imgs[0]))[0]
    corresponding_csv = os.path.join(nestin_csv_dir, first_image_name + '.csv')
    if os.path.exists(corresponding_csv):
        plot_sample_with_dots(nestin_imgs[0], corresponding_csv)

# ✅ Dot CSV → Binary Mask with configurable radius
def dots_to_mask(csv_file, shape=(600,800), radius=DISK_RADIUS):
    mask = np.zeros(shape, dtype=np.uint8)
    coords = pd.read_csv(csv_file)
    for _, row in coords.iterrows():
        rr, cc = disk((int(row['Y']), int(row['X'])), radius=radius, shape=shape)
        mask[rr, cc] = 1
    return mask

# ✅ Generate Masks
image_folders = sorted(glob(os.path.join(IDCIA_PATH, "IDCIA/images/*")))
mask_dir = os.path.join(OUTPUT_PATH, "generated_masks")
os.makedirs(mask_dir, exist_ok=True)

image_paths, mask_paths, antibody_types, image_names = [], [], [], []
for folder in image_folders:
    antibody = os.path.basename(folder)
    save_folder = os.path.join(mask_dir, antibody)
    os.makedirs(save_folder, exist_ok=True)
    for image_file in glob(os.path.join(folder, '*.tiff')):
        fname = os.path.splitext(os.path.basename(image_file))[0]
        csv_file = os.path.join(IDCIA_PATH, 'IDCIA/ground_truth', antibody, fname + '.csv')
        if os.path.exists(csv_file):
            mask = dots_to_mask(csv_file, radius=DISK_RADIUS)
            save_path = os.path.join(save_folder, fname + '.npy')
            np.save(save_path, mask)

            png_path = os.path.join(save_folder, fname + '.png')
            cv2.imwrite(png_path, mask * 255)

            image_paths.append(image_file)
            mask_paths.append(save_path)
            antibody_types.append(antibody)
            image_names.append(fname)

print(f"\nFinal usable image-mask pairs: {len(image_paths)}")

# ✅ Segmentation Dataset with Augmentation
class IDCIA_Dataset(Dataset):
    def __init__(self, image_paths, mask_paths, antibody_types=None, image_names=None, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.antibody_types = antibody_types
        self.image_names = image_names
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load image and convert to float
        img = np.array(Image.open(self.image_paths[idx])).astype(np.float32)

        # Proper normalization pipeline
        img = img / 255.0
        mean = img.mean()
        std = img.std() + 1e-7
        img = (img - mean) / std

        # Load mask
        mask = np.load(self.mask_paths[idx])

        # Convert to tensors
        img = torch.from_numpy(img).unsqueeze(0).float()
        mask = torch.from_numpy(mask).unsqueeze(0).float()

        # Apply transformations if provided
        if self.transform:
            combined = torch.cat([img, mask], dim=0)
            combined = self.transform(combined)
            img, mask = combined[0].unsqueeze(0), combined[1].unsqueeze(0)

        # Return with metadata if available
        if self.antibody_types and self.image_names:
            return img, mask, self.antibody_types[idx], self.image_names[idx]
        return img, mask

# ✅ Enhanced Augmentations
train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(180),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1), scale=(0.9, 1.1)),
    transforms.RandomResizedCrop((IMG_SIZE, IMG_SIZE), scale=(0.8, 1.2)),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.33), ratio=(0.3, 3.3)),
    transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 2.0)),
])

# Minimal transformations for validation
val_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
])

# Train-val split
dataset = IDCIA_Dataset(image_paths, mask_paths, antibody_types, image_names, transform=None)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

train_dataset.dataset.transform = train_transform
val_dataset.dataset.transform = val_transform

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)

# ✅ Attention Gate Module for segmentation
class AttentionGate(nn.Module):
    def __init__(self, F_g, F_l, F_int):
        super(AttentionGate, self).__init__()
        self.W_g = nn.Sequential(
            nn.Conv2d(F_g, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )
        self.W_x = nn.Sequential(
            nn.Conv2d(F_l, F_int, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(F_int)
        )
        self.psi = nn.Sequential(
            nn.Conv2d(F_int, 1, kernel_size=1, stride=1, padding=0, bias=True),
            nn.BatchNorm2d(1),
            nn.Sigmoid()
        )
        self.relu = nn.ReLU(inplace=True)

    def forward(self, g, x):
        g1 = self.W_g(g)
        x1 = self.W_x(x)
        psi = self.relu(g1 + x1)
        psi = self.psi(psi)
        return x * psi

# ✅ Enhanced UNet++ Model with Attention Gates
class AttentionUNetPlusPlus(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, base_channels=64, use_attention=True):
        super().__init__()
        self.use_attention = use_attention
        self.use_checkpointing = True

        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, padding=1),
                nn.BatchNorm2d(out_c),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, 3, padding=1),
                nn.BatchNorm2d(out_c),
                nn.ReLU(inplace=True)
            )

        # Base channels
        bc = base_channels
        self.conv00 = conv_block(in_channels, bc)
        self.conv10 = conv_block(bc, bc * 2)
        self.conv20 = conv_block(bc * 2, bc * 4)
        self.conv30 = conv_block(bc * 4, bc * 8)
        self.conv40 = conv_block(bc * 8, bc * 16)

        # Attention gates for upsampling paths
        if use_attention:
            self.att1 = AttentionGate(F_g=bc*2, F_l=bc, F_int=bc//2)
            self.att2 = AttentionGate(F_g=bc*4, F_l=bc*2, F_int=bc)
            self.att3 = AttentionGate(F_g=bc*8, F_l=bc*4, F_int=bc*2)
            self.att4 = AttentionGate(F_g=bc*16, F_l=bc*8, F_int=bc*4)

            # Nested attention gates
            self.att12 = AttentionGate(F_g=bc*2, F_l=bc, F_int=bc//2)
            self.att22 = AttentionGate(F_g=bc*4, F_l=bc*2, F_int=bc)
            self.att32 = AttentionGate(F_g=bc*8, F_l=bc*4, F_int=bc*2)

            self.att13 = AttentionGate(F_g=bc*2, F_l=bc, F_int=bc//2)
            self.att23 = AttentionGate(F_g=bc*4, F_l=bc*2, F_int=bc)

            self.att14 = AttentionGate(F_g=bc*2, F_l=bc, F_int=bc//2)

        self.up01 = conv_block(bc + bc * 2, bc)
        self.up11 = conv_block(bc * 2 + bc * 4, bc * 2)
        self.up21 = conv_block(bc * 4 + bc * 8, bc * 4)
        self.up31 = conv_block(bc * 8 + bc * 16, bc * 8)

        self.up02 = conv_block(bc * 2 + bc * 2, bc)
        self.up12 = conv_block(bc * 4 + bc * 4, bc * 2)
        self.up22 = conv_block(bc * 8 + bc * 8, bc * 4)

        self.up03 = conv_block(bc * 3 + bc * 2, bc)
        self.up13 = conv_block(bc * 6 + bc * 4, bc * 2)

        self.up04 = conv_block(bc * 4 + bc * 2, bc)

        self.final = nn.Conv2d(bc, out_channels, kernel_size=1)
        self.maxpool = nn.MaxPool2d(2)
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)

    def forward(self, x):
        # Enable checkpointing of the encoder blocks to save memory
        if self.use_checkpointing and self.training:
            x00 = torch.utils.checkpoint.checkpoint(self.conv00, x)
            x10 = torch.utils.checkpoint.checkpoint(self.conv10, self.maxpool(x00))
            x20 = torch.utils.checkpoint.checkpoint(self.conv20, self.maxpool(x10))
            x30 = torch.utils.checkpoint.checkpoint(self.conv30, self.maxpool(x20))
            x40 = torch.utils.checkpoint.checkpoint(self.conv40, self.maxpool(x30))
        else:
            x00 = self.conv00(x)
            x10 = self.conv10(self.maxpool(x00))
            x20 = self.conv20(self.maxpool(x10))
            x30 = self.conv30(self.maxpool(x20))
            x40 = self.conv40(self.maxpool(x30))

        # First level nested path with attention
        if self.use_attention:
            x10_up = self.upsample(x10)
            x00_att = self.att1(x10_up, x00)
            x01 = self.up01(torch.cat([x00_att, x10_up], 1))

            x20_up = self.upsample(x20)
            x10_att = self.att2(x20_up, x10)
            x11 = self.up11(torch.cat([x10_att, x20_up], 1))

            x30_up = self.upsample(x30)
            x20_att = self.att3(x30_up, x20)
            x21 = self.up21(torch.cat([x20_att, x30_up], 1))

            x40_up = F.interpolate(x40, size=x30.shape[2:], mode='bilinear', align_corners=True)
            x30_att = self.att4(x40_up, x30)
            x31 = self.up31(torch.cat([x30_att, x40_up], dim=1))
        else:
            x01 = self.up01(torch.cat([x00, self.upsample(x10)], 1))
            x11 = self.up11(torch.cat([x10, self.upsample(x20)], 1))
            x21 = self.up21(torch.cat([x20, self.upsample(x30)], 1))
            x40_up = F.interpolate(x40, size=x30.shape[2:], mode='bilinear', align_corners=True)
            x31 = self.up31(torch.cat([x30, x40_up], dim=1))

        # Second level nested path with attention
        if self.use_attention:
            x11_up = self.upsample(x11)
            x00_att2 = self.att12(x11_up, x00)
            x01_att2 = self.att12(x11_up, x01)
            x02 = self.up02(torch.cat([x00_att2, x01_att2, x11_up], 1))

            x21_up = self.upsample(x21)
            x10_att2 = self.att22(x21_up, x10)
            x11_att2 = self.att22(x21_up, x11)
            x12 = self.up12(torch.cat([x10_att2, x11_att2, x21_up], 1))

            x31_up = self.upsample(x31)
            x20_att2 = self.att32(x31_up, x20)
            x21_att2 = self.att32(x31_up, x21)
            x22 = self.up22(torch.cat([x20_att2, x21_att2, x31_up], 1))
        else:
            x02 = self.up02(torch.cat([x00, x01, self.upsample(x11)], 1))
            x12 = self.up12(torch.cat([x10, x11, self.upsample(x21)], 1))
            x22 = self.up22(torch.cat([x20, x21, self.upsample(x31)], 1))

        # Third level nested path
        if self.use_attention:
            x12_up = self.upsample(x12)
            x00_att3 = self.att13(x12_up, x00)
            x01_att3 = self.att13(x12_up, x01)
            x02_att3 = self.att13(x12_up, x02)
            x03 = self.up03(torch.cat([x00_att3, x01_att3, x02_att3, x12_up], 1))

            x22_up = self.upsample(x22)
            x10_att3 = self.att23(x22_up, x10)
            x11_att3 = self.att23(x22_up, x11)
            x12_att3 = self.att23(x22_up, x12)
            x13 = self.up13(torch.cat([x10_att3, x11_att3, x12_att3, x22_up], 1))
        else:
            x03 = self.up03(torch.cat([x00, x01, x02, self.upsample(x12)], 1))
            x13 = self.up13(torch.cat([x10, x11, x12, self.upsample(x22)], 1))

        # Final nested path
        if self.use_attention:
            x13_up = self.upsample(x13)
            x00_att4 = self.att14(x13_up, x00)
            x01_att4 = self.att14(x13_up, x01)
            x02_att4 = self.att14(x13_up, x02)
            x03_att4 = self.att14(x13_up, x03)
            x04 = self.up04(torch.cat([x00_att4, x01_att4, x02_att4, x03_att4, x13_up], 1))
        else:
            x04 = self.up04(torch.cat([x00, x01, x02, x03, self.upsample(x13)], 1))

        # Output layer
        output = self.final(x04)
        return output

# ✅ Original UNet++ model class for loading previous weights
class UNetPlusPlus(nn.Module):
    def __init__(self, in_channels=1, out_channels=1, base_channels=64):
        super().__init__()

        def conv_block(in_c, out_c):
            return nn.Sequential(
                nn.Conv2d(in_c, out_c, 3, padding=1),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_c, out_c, 3, padding=1),
                nn.ReLU(inplace=True)
            )

        bc = base_channels
        self.conv00 = conv_block(in_channels, bc)
        self.conv10 = conv_block(bc, bc * 2)
        self.conv20 = conv_block(bc * 2, bc * 4)
        self.conv30 = conv_block(bc * 4, bc * 8)
        self.conv40 = conv_block(bc * 8, bc * 16)

        self.up01 = conv_block(bc + bc * 2, bc)
        self.up11 = conv_block(bc * 2 + bc * 4, bc * 2)
        self.up21 = conv_block(bc * 4 + bc * 8, bc * 4)
        self.up31 = conv_block(bc * 8 + bc * 16, bc * 8)

        self.up02 = conv_block(bc * 2 + bc * 2, bc)
        self.up12 = conv_block(bc * 4 + bc * 4, bc * 2)
        self.up22 = conv_block(bc * 8 + bc * 8, bc * 4)

        self.up03 = conv_block(bc * 3 + bc * 2, bc)
        self.up13 = conv_block(bc * 6 + bc * 4, bc * 2)

        self.up04 = conv_block(bc * 4 + bc * 2, bc)

        self.final = nn.Conv2d(bc, out_channels, kernel_size=1)
        self.maxpool = nn.MaxPool2d(2)
        self.upsample = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)

    def forward(self, x):
        x00 = self.conv00(x)
        x10 = self.conv10(self.maxpool(x00))
        x20 = self.conv20(self.maxpool(x10))
        x30 = self.conv30(self.maxpool(x20))
        x40 = self.conv40(self.maxpool(x30))

        x01 = self.up01(torch.cat([x00, self.upsample(x10)], 1))
        x11 = self.up11(torch.cat([x10, self.upsample(x20)], 1))
        x21 = self.up21(torch.cat([x20, self.upsample(x30)], 1))
        x40_up = F.interpolate(x40, size=x30.shape[2:], mode='bilinear', align_corners=True)
        x31 = self.up31(torch.cat([x30, x40_up], dim=1))

        x02 = self.up02(torch.cat([x00, x01, self.upsample(x11)], 1))
        x12 = self.up12(torch.cat([x10, x11, self.upsample(x21)], 1))
        x22 = self.up22(torch.cat([x20, x21, self.upsample(x31)], 1))

        x03 = self.up03(torch.cat([x00, x01, x02, self.upsample(x12)], 1))
        x13 = self.up13(torch.cat([x10, x11, x12, self.upsample(x22)], 1))

        x04 = self.up04(torch.cat([x00, x01, x02, x03, self.upsample(x13)], 1))

        output = self.final(x04)
        return output

# ✅ Loss function for segmentation
class WeightedDiceBCEWithLogitsLoss(nn.Module):
    def __init__(self, pos_weight=20.0):
        super().__init__()
        self.pos_weight = torch.tensor([pos_weight])
        self.bce = None

    def forward(self, pred, target):
        if self.bce is None:
            self.pos_weight = self.pos_weight.to(pred.device)
            self.bce = nn.BCEWithLogitsLoss(pos_weight=self.pos_weight)

        bce_loss = self.bce(pred, target)

        pred_sigmoid = torch.sigmoid(pred)

        smooth = 1e-5
        intersection = (pred_sigmoid * target).sum()
        dice_loss = 1 - (2. * intersection + smooth) / (pred_sigmoid.sum() + target.sum() + smooth)

        return 0.7 * bce_loss + 0.3 * dice_loss

# ✅ Load pretrained model function
def load_best_model(model):
    standard_path = os.path.join(OUTPUT_PATH, "best_model.pth")
    input_path = os.path.join(MODEL_PATH, "best_model.pth")

    if os.path.exists(input_path):
        print(f"Loading pre-trained model from dataset input: {input_path}")
        best_model_path = input_path
    elif os.path.exists(standard_path):
        print(f"Loading pre-trained model from working directory: {standard_path}")
        best_model_path = standard_path
    else:
        print("No pre-trained model found, starting with fresh weights.")
        return False

    try:
        checkpoint = torch.load(best_model_path, weights_only=True)
        model.load_state_dict(checkpoint, strict=False)
        print("Model loaded successfully!")
        return True
    except Exception as e:
        print(f"Error loading model directly: {e}")
        print("Creating compatibility layer for model architecture differences...")

        device = next(model.parameters()).device
        temp_model = UNetPlusPlus(in_channels=1, out_channels=1, base_channels=48).to(device)

        try:
            temp_model.load_state_dict(torch.load(best_model_path, weights_only=True))
            print("Successfully loaded weights into temporary model")

            model_dict = model.state_dict()

            for name, param in temp_model.state_dict().items():
                if name in model_dict and model_dict[name].shape == param.shape:
                    model_dict[name].copy_(param)
                    print(f"Copied weights for: {name}")

            print("Transferred compatible weights from regular UNet++ to Attention UNet++")
            del temp_model
            torch.cuda.empty_cache()
            return True

        except Exception as inner_e:
            print(f"Failed to load model weights: {inner_e}")
            return False

# ✅ Test-Time Augmentation function
def tta_predict(model, img_tensor, device):
    """Memory-optimized test-time augmentation"""
    model.eval()

    with torch.no_grad():
        # Original prediction
        if USE_MIXED_PRECISION:
            with torch.amp.autocast('cuda'):
                pred_orig = model(img_tensor)
        else:
            pred_orig = model(img_tensor)

        # Horizontal flip
        img_h = torch.flip(img_tensor, dims=[3])
        if USE_MIXED_PRECISION:
            with torch.amp.autocast('cuda'):
                pred_h = model(img_h)
        else:
            pred_h = model(img_h)
        pred_h = torch.flip(pred_h, dims=[3])

        # Add results and free memory
        pred_combined = pred_orig + pred_h
        del pred_orig, pred_h
        torch.cuda.empty_cache()

        # Vertical flip
        img_v = torch.flip(img_tensor, dims=[2])
        if USE_MIXED_PRECISION:
            with torch.amp.autocast('cuda'):
                pred_v = model(img_v)
        else:
            pred_v = model(img_v)
        pred_v = torch.flip(pred_v, dims=[2])

        # Final combination
        pred_combined = (pred_combined + pred_v) / 3.0
        del pred_v
        torch.cuda.empty_cache()

    return pred_combined

# ✅ Initialize segmentation model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = AttentionUNetPlusPlus(
    in_channels=1,
    out_channels=1,
    base_channels=BASE_CHANNELS,
    use_attention=USE_ATTENTION
).to(device)
criterion = WeightedDiceBCEWithLogitsLoss(pos_weight=20.0)

# ✅ Enhanced Post-processing function
def postprocess_mask(mask, threshold=0.5, min_size=30, max_size=1000):
    """Apply morphological operations to clean up the mask"""

    # Threshold
    binary = (mask > threshold).astype(np.uint8)

    # Remove small objects
    labeled, num_features = ndimage.label(binary)
    if num_features > 0:  # Only process if there are objects
        sizes = ndimage.sum(binary, labeled, range(1, num_features + 1))

        # Remove small objects
        mask_sizes = sizes < min_size
        remove_pixels = mask_sizes[labeled - 1]
        binary[remove_pixels] = 0

        # Remove very large objects (likely artifacts)
        if max_size > 0:
            mask_sizes = sizes > max_size
            remove_pixels = mask_sizes[labeled - 1]
            binary[remove_pixels] = 0

    # Dilate to connect nearby components
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3, 3))
    binary = cv2.dilate(binary, kernel, iterations=1)

    # Fill holes in detected cells
    for i in range(1, num_features + 1):
        component = (labeled == i).astype(np.uint8)
        filled = cv2.morphologyEx(component, cv2.MORPH_CLOSE, kernel)
        binary[filled > 0] = 1

    return binary

# ✅ TRAIN SEGMENTATION MODEL (since LOAD_PRETRAINED = False)
print("\nTraining segmentation model from scratch...")
optimizer = torch.optim.AdamW(model.parameters(), lr=SEG_LEARNING_RATE, weight_decay=1e-4)
scheduler = CosineAnnealingWarmRestarts(
    optimizer,
    T_0=7,
    T_mult=2,
    eta_min=5e-7
)

# Mixed precision setup
if USE_MIXED_PRECISION:
    scaler = torch.amp.GradScaler('cuda') if hasattr(torch.amp, 'GradScaler') else torch.cuda.amp.GradScaler()
    print("Using mixed precision training for memory optimization")

# Training Loop
best_val_loss = np.inf
patience_counter = 0
early_stopping_patience = 20

for epoch in range(1, NUM_EPOCHS + 1):
    model.train()
    train_loss = 0.0
    for x, y, *_ in train_loader:
        x, y = x.to(device), y.to(device)
        optimizer.zero_grad()

        # Use mixed precision with updated API
        if USE_MIXED_PRECISION:
            with torch.amp.autocast('cuda'):
                preds = model(x)
                loss = criterion(preds, y)  # Calculate loss inside autocast

            # Handle mixed precision backward
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            preds = model(x)
            loss = criterion(preds, y)
            loss.backward()
            optimizer.step()

        train_loss += loss.item() * x.size(0)

        # Free memory explicitly
        del preds
        torch.cuda.empty_cache()

    # Calculate average training loss correctly
    train_loss /= len(train_loader.dataset)

    # Clear cache between training and validation
    torch.cuda.empty_cache()

    # Validation phase
    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for x, y, *_ in val_loader:
            x, y = x.to(device), y.to(device)

            if USE_MIXED_PRECISION:
                with torch.amp.autocast('cuda'):
                    preds = model(x)
                    loss = criterion(preds, y)
            else:
                preds = model(x)
                loss = criterion(preds, y)

            val_loss += loss.item() * x.size(0)

    # Calculate average validation loss correctly
    val_loss /= len(val_loader.dataset)

    print(f"[Epoch {epoch}/{NUM_EPOCHS}] Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")

    # Update scheduler
    scheduler.step()

    # Clear cache after validation
    torch.cuda.empty_cache()

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        patience_counter = 0
        torch.save(model.state_dict(), os.path.join(OUTPUT_PATH, "best_model.pth"))
        print(f"--> Saved best model at Epoch {epoch}")
    else:
        patience_counter += 1
        if patience_counter >= early_stopping_patience:
            print("Early stopping triggered!")
            break

# Load the best model for inference
print("\nLoading best model for inference...")
model.load_state_dict(torch.load(os.path.join(OUTPUT_PATH, "best_model.pth")))
model.eval()

# ✅ Memory-efficient threshold evaluation with TTA
def evaluate_thresholds(model, dataloader):
    thresholds = np.linspace(0.2, 0.7, 20)  # Much lower than current 0.9
    model.eval()
    threshold_scores = []

    with torch.no_grad():
        for threshold in thresholds:
            dices = []
            for x, y, *_ in dataloader:
                x, y = x.to(device), y.to(device)

                if USE_TTA:
                    # Use TTA for improved predictions
                    preds_logits = tta_predict(model, x, device)
                    preds = torch.sigmoid(preds_logits)
                else:
                    # Regular prediction
                    if USE_MIXED_PRECISION:
                        with torch.amp.autocast('cuda'):
                            preds_logits = model(x)
                            preds = torch.sigmoid(preds_logits)
                    else:
                        preds_logits = model(x)
                        preds = torch.sigmoid(preds_logits)

                # Threshold probabilities
                preds = (preds > threshold).float()

                # Calculate Dice for each image
                for p, t in zip(preds, y):
                    intersection = (p * t).sum().item()
                    union = p.sum().item() + t.sum().item()
                    dice = (2 * intersection) / union if union != 0 else 1.0
                    dices.append(dice)

            mean_dice = np.mean(dices)
            threshold_scores.append((threshold, mean_dice))
            print(f"Threshold: {threshold:.4f} | Mean Dice: {mean_dice:.4f}")

        # Free memory after each threshold evaluation
        torch.cuda.empty_cache()

    # Plot threshold performance
    plot_threshold_performance(threshold_scores)

    return threshold_scores

# ✅ Added threshold visualization
def plot_threshold_performance(threshold_scores):
    thresholds, scores = zip(*threshold_scores)
    plt.figure(figsize=(12, 6))
    plt.plot(thresholds, scores, 'o-')
    plt.xlabel('Threshold Value')
    plt.ylabel('Mean Dice Score')
    plt.title('Threshold vs. Performance')
    plt.grid(True)

    # Find and mark the best threshold
    best_threshold, best_score = max(threshold_scores, key=lambda x: x[1])
    plt.axvline(x=best_threshold, color='r', linestyle='--')
    plt.axhline(y=best_score, color='r', linestyle='--')
    plt.scatter([best_threshold], [best_score], color='red', s=100, zorder=5)
    plt.annotate(f'Best: {best_threshold:.4f}, {best_score:.4f}',
                 xy=(best_threshold, best_score),
                 xytext=(best_threshold+0.1, best_score-0.05),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    plt.tight_layout()
    plt.show()

print("\nEvaluating thresholds to find optimal value...")
threshold_scores = evaluate_thresholds(model, val_loader)
optimal_threshold, optimal_dice = max(threshold_scores, key=lambda x: x[1])
print(f"\nOptimal Threshold: {optimal_threshold:.4f} with Mean Dice: {optimal_dice:.4f}")

# Add after optimal threshold determination and before mask generation for classification

# Import seg_metrics for Hausdorff distance calculation
from seg_metrics.seg_metrics import write_metrics

# ✅ Hausdorff distance computation
def compute_hd95(pred_mask, true_mask):
    pred_mask = (pred_mask > 0).astype(np.uint8)
    true_mask = (true_mask > 0).astype(np.uint8)
    metrics = write_metrics(
        labels=[1],
        gdth_img=true_mask,
        pred_img=pred_mask,
        csv_file=None,
        metrics=['hd95']
    )
    return metrics[0]['hd95'][0]

# ✅ IoU + Dice + Hausdorff Evaluation
def compute_metrics(pred, target, threshold):
    # Apply sigmoid to get probabilities
    pred = torch.sigmoid(pred)
    # Then threshold
    pred = (pred > threshold).float()

    intersection = (pred * target).sum().item()
    union = ((pred + target) > 0).float().sum().item()
    pred_sum = pred.sum().item()
    target_sum = target.sum().item()

    iou = intersection / union if union != 0 else 1.0
    dice = (2 * intersection) / (pred_sum + target_sum) if (pred_sum + target_sum) != 0 else 1.0

    pred_np = pred.squeeze().cpu().numpy()
    target_np = target.squeeze().cpu().numpy()
    hd95 = compute_hd95(pred_np, target_np)

    return iou, dice, hd95

# ✅ Memory-efficient final evaluation with TTA
ious, dices, hd95s = [], [], []
model.eval()
print("\nComputing final metrics...")
with torch.no_grad():
    for x, y, *_ in val_loader:
        x, y = x.to(device), y.to(device)

        if USE_TTA:
            # Use TTA for improved predictions
            preds_logits = tta_predict(model, x, device)
        else:
            # Regular prediction
            if USE_MIXED_PRECISION:
                with autocast():
                    preds_logits = model(x)
            else:
                preds_logits = model(x)

        for p, t in zip(preds_logits, y):
            i, d, h = compute_metrics(p.unsqueeze(0), t.unsqueeze(0), threshold=optimal_threshold)
            ious.append(i)
            dices.append(d)
            hd95s.append(h)

print(f"\nMean IoU: {np.mean(ious):.4f}, Dice: {np.mean(dices):.4f}, HD95: {np.mean(hd95s):.4f}")

# ✅ Save metrics to CSV file
def save_metrics_to_csv():
    metrics_df = pd.DataFrame({
        'IoU': ious,
        'Dice': dices,
        'HD95': hd95s
    })
    metrics_df.loc['Mean'] = metrics_df.mean()
    metrics_df.to_csv(os.path.join(OUTPUT_PATH, 'validation_metrics.csv'), index=False)
    print(f"Saved validation metrics to {os.path.join(OUTPUT_PATH, 'validation_metrics.csv')}")

save_metrics_to_csv()

# ✅ Generate segmentation masks for classification
def generate_segmentation_masks(image_paths, seg_model, batch_size=4):
    """Pre-compute segmentation masks for all images in batches"""
    masks = []
    seg_model.eval()

    # Create a simple dataset for batch processing
    class ImageDataset(Dataset):
        def __init__(self, image_paths):
            self.image_paths = image_paths

        def __len__(self):
            return len(self.image_paths)

        def __getitem__(self, idx):
            img_path = self.image_paths[idx]
            img = np.array(Image.open(img_path)).astype(np.float32)
            img = img / 255.0
            mean = img.mean()
            std = img.std() + 1e-7
            img = (img - mean) / std
            return torch.from_numpy(img).unsqueeze(0).float(), img_path

    # Create dataloader for batch processing
    image_dataset = ImageDataset(image_paths)
    dataloader = DataLoader(image_dataset, batch_size=batch_size, shuffle=False)

    print("Pre-computing segmentation masks...")
    mask_dict = {}

    with torch.no_grad():
        for batch_imgs, batch_paths in dataloader:
            batch_imgs = batch_imgs.to(device)

            # Generate masks
            if USE_TTA:
                pred_logits = tta_predict(seg_model, batch_imgs, device)
            else:
                pred_logits = seg_model(batch_imgs)

            pred_probs = torch.sigmoid(pred_logits)

            # Process each image in the batch
            for i, path in enumerate(batch_paths):
                pred_mask = pred_probs[i].squeeze().cpu().numpy()
                proc_mask = postprocess_mask(pred_mask, threshold=optimal_threshold,
                                           min_size=30, max_size=1000)
                mask_dict[path] = proc_mask

    # Ensure masks are in the same order as image_paths
    masks = [mask_dict[path] for path in image_paths]
    print(f"Generated {len(masks)} masks")
    return masks

# ✅ Create dataset for classification using segmentation masks
class IDCIA_MaskedClassifierDataset(Dataset):
    def __init__(self, image_paths, masks, labels, transform=None):
        self.image_paths = image_paths
        self.masks = masks
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Load original image
        original_img = np.array(Image.open(self.image_paths[idx])).astype(np.float32)
        original_img_norm = original_img / 255.0

        # Apply pre-computed mask
        mask = self.masks[idx]

        # Create 3-channel input: original, mask, masked image
        masked_img = original_img_norm * mask

        # Stack as channels
        combined = np.stack([
            original_img_norm,  # Original image
            mask,               # Binary mask
            masked_img          # Masked image
        ])

        # Convert to tensor
        img_tensor = torch.from_numpy(combined).float()

        # Resize for classification
        img_tensor = F.interpolate(img_tensor.unsqueeze(0), size=(128, 128),
                                  mode='bilinear', align_corners=True).squeeze(0)

        if self.transform:
            img_tensor = self.transform(img_tensor)

        return img_tensor, self.labels[idx]

# ✅ Generate labels for classification
antibody_to_idx = {os.path.basename(p): i for i, p in enumerate(image_folders)}
labels = [antibody_to_idx[os.path.basename(os.path.dirname(p))] for p in image_paths]

# ✅ Generate segmentation masks for all images
print("\nGenerating segmentation masks for classification...")
segmentation_masks = generate_segmentation_masks(image_paths, model, batch_size=4)

# ✅ Create transform for 3-channel input
transform = transforms.Compose([
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Create the masked dataset for classification
masked_cls_dataset = IDCIA_MaskedClassifierDataset(
    image_paths=image_paths,
    masks=segmentation_masks,
    labels=labels,
    transform=transform
)

# Split dataset for training/validation
cls_train_size = int(0.8 * len(masked_cls_dataset))
cls_val_size = len(masked_cls_dataset) - cls_train_size
masked_train_dataset, masked_val_dataset = random_split(masked_cls_dataset, [cls_train_size, cls_val_size])

# Create data loaders for classification
masked_train_loader = DataLoader(masked_train_dataset, batch_size=8, shuffle=True)
masked_val_loader = DataLoader(masked_val_dataset, batch_size=8, shuffle=False)

# ✅ Simple CNN Classifier
class MaskedClassifierCNN(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),  # Changed from 1 to 3 input channels
            nn.Conv2d(16, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x

# ✅ Initialize and train CNN classifier
masked_cnn = MaskedClassifierCNN(num_classes=len(antibody_to_idx)).to(device)
criterion_cnn = nn.CrossEntropyLoss()
optimizer_cnn = torch.optim.Adam(masked_cnn.parameters(), lr=CLS_LEARNING_RATE)

print("\nTraining CNN classifier on segmented regions:")
for epoch in range(NUM_EPOCHS):
    # Training phase
    masked_cnn.train()
    train_correct, train_total = 0, 0

    for x, y in masked_train_loader:
        x, y = x.to(device), y.to(device)
        out = masked_cnn(x)
        loss = criterion_cnn(out, y)

        optimizer_cnn.zero_grad()
        loss.backward()
        optimizer_cnn.step()

        preds = out.argmax(dim=1)
        train_correct += (preds == y).sum().item()
        train_total += y.size(0)

    # Validation phase
    masked_cnn.eval()
    val_correct, val_total = 0, 0

    with torch.no_grad():
        for x, y in masked_val_loader:
            x, y = x.to(device), y.to(device)
            out = masked_cnn(x)
            preds = out.argmax(dim=1)
            val_correct += (preds == y).sum().item()
            val_total += y.size(0)

    train_acc = train_correct / train_total
    val_acc = val_correct / val_total
    print(f"[CNN Epoch {epoch+1}/{NUM_EPOCHS}] Train Acc: {train_acc*100:.2f}%, Val Acc: {val_acc*100:.2f}%")

# ✅ Evaluate CNN and create confusion matrix
masked_cnn.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for x, y in masked_val_loader:
        x = x.to(device)
        out = masked_cnn(x)
        preds = out.argmax(dim=1).cpu().numpy()
        y_true.extend(y.numpy())
        y_pred.extend(preds)

cm = confusion_matrix(y_true, y_pred, labels=range(len(antibody_to_idx)))
plt.figure(figsize=(12, 10))
antibody_names = list(antibody_to_idx.keys())
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=antibody_names).plot(xticks_rotation=45, cmap='Blues')
plt.title("Confusion Matrix: CNN with Segmentation Masking")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, "masked_cnn_confusion_matrix.png"))
plt.show()

# ✅ ResNet50-based classifier with 3-channel input
resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
# Modify for 3-channel input (original, mask, masked)
resnet.conv1 = nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False)
resnet.fc = nn.Linear(resnet.fc.in_features, len(antibody_to_idx))
resnet = resnet.to(device)

criterion_resnet = nn.CrossEntropyLoss()
optimizer_resnet = torch.optim.Adam(resnet.parameters(), lr=CLS_LEARNING_RATE * 0.5)  # Lower LR for ResNet50

print("\nTraining ResNet50 classifier on segmented regions:")
for epoch in range(NUM_EPOCHS):  # Running for 20 epochs as requested
    # Training phase
    resnet.train()
    train_correct, train_total = 0, 0

    for x, y in masked_train_loader:
        x, y = x.to(device), y.to(device)
        out = resnet(x)
        loss = criterion_resnet(out, y)

        optimizer_resnet.zero_grad()
        loss.backward()
        optimizer_resnet.step()

        preds = out.argmax(dim=1)
        train_correct += (preds == y).sum().item()
        train_total += y.size(0)

    # Validation phase
    resnet.eval()
    val_correct, val_total = 0, 0

    with torch.no_grad():
        for x, y in masked_val_loader:
            x, y = x.to(device), y.to(device)
            out = resnet(x)
            preds = out.argmax(dim=1)
            val_correct += (preds == y).sum().item()
            val_total += y.size(0)

    train_acc = train_correct / train_total
    val_acc = val_correct / val_total
    print(f"[ResNet50 Epoch {epoch+1}/{NUM_EPOCHS}] Train Acc: {train_acc*100:.2f}%, Val Acc: {val_acc*100:.2f}%")

# Evaluate ResNet50 and create confusion matrix
resnet.eval()
y_true, y_pred = [], []

with torch.no_grad():
    for x, y in masked_val_loader:
        x = x.to(device)
        out = resnet(x)
        preds = out.argmax(dim=1).cpu().numpy()
        y_true.extend(y.numpy())
        y_pred.extend(preds)

cm = confusion_matrix(y_true, y_pred, labels=range(len(antibody_to_idx)))
plt.figure(figsize=(12, 10))
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=antibody_names).plot(xticks_rotation=45, cmap='Greens')
plt.title("Confusion Matrix: ResNet50 with Segmentation Masking")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, "masked_resnet_confusion_matrix.png"))
plt.show()

# Print detailed classification metrics
print("\nClassification Report (ResNet50 with segmentation masking):")
print(classification_report(y_true, y_pred, target_names=antibody_names))

# ✅ Visualize examples of masked images
def visualize_masked_examples():
    plt.figure(figsize=(15, 10))
    for i in range(min(5, len(masked_cls_dataset))):
        # Get original image
        original_img = np.array(Image.open(image_paths[i])).astype(np.float32) / 255.0

        # Get mask and masked image
        mask = segmentation_masks[i]
        masked_img = original_img * mask

        # Display images
        plt.subplot(5, 3, i*3 + 1)
        plt.imshow(original_img, cmap='gray')
        plt.title(f"Original ({antibody_types[i]})")
        plt.axis('off')

        plt.subplot(5, 3, i*3 + 2)
        plt.imshow(mask, cmap='gray')
        plt.title("Segmentation Mask")
        plt.axis('off')

        plt.subplot(5, 3, i*3 + 3)
        plt.imshow(masked_img, cmap='gray')
        plt.title("Masked Image")
        plt.axis('off')

    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_PATH, "masked_examples.png"))
    plt.show()

# Show examples of the masked images used for classification
visualize_masked_examples()

# ✅ Compare performance between CNN and ResNet50
def compare_classifier_performance():
    # Get predictions from both models
    masked_cnn.eval()
    resnet.eval()

    all_labels = []
    cnn_preds = []
    resnet_preds = []

    with torch.no_grad():
        for x, y in masked_val_loader:
            x = x.to(device)

            # CNN predictions
            out_cnn = masked_cnn(x)
            pred_cnn = out_cnn.argmax(dim=1).cpu().numpy()

            # ResNet predictions
            out_resnet = resnet(x)
            pred_resnet = out_resnet.argmax(dim=1).cpu().numpy()

            # Store results
            all_labels.extend(y.numpy())
            cnn_preds.extend(pred_cnn)
            resnet_preds.extend(pred_resnet)

    # Calculate accuracies
    cnn_acc = np.mean(np.array(cnn_preds) == np.array(all_labels))
    resnet_acc = np.mean(np.array(resnet_preds) == np.array(all_labels))

    # Plot comparison
    plt.figure(figsize=(10, 6))
    models = ['CNN', 'ResNet50']
    accuracies = [cnn_acc * 100, resnet_acc * 100]

    plt.bar(models, accuracies, color=['skyblue', 'lightgreen'])
    plt.xlabel('Model Architecture')
    plt.ylabel('Validation Accuracy (%)')
    plt.title('Comparison of Classifier Performance on Segmented Cells')
    plt.ylim([0, 100])

    # Add accuracy values on bars
    for i, acc in enumerate(accuracies):
        plt.text(i, acc + 1, f"{acc:.2f}%", ha='center')

    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_PATH, "classifier_comparison.png"))
    plt.show()

    # Print class-wise performance for the better model
    if resnet_acc > cnn_acc:
        print("\nDetailed ResNet50 Performance (Better Model):")
        print(classification_report(all_labels, resnet_preds, target_names=antibody_names))
    else:
        print("\nDetailed CNN Performance (Better Model):")
        print(classification_report(all_labels, cnn_preds, target_names=antibody_names))

# Compare the performance of both classifiers
compare_classifier_performance()

print("\nComplete IDCIA analysis finished successfully!")
```
